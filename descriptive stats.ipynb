```python
ig = pd.read_csv("data/clean_latam.csv", usecols = ['ctry', 'year', 'Intermediate goods_tariff', 'Capital goods_tariff',
                                                   'Consumer goods_tariff', 'Raw materials_tariff'])

ig = ig.loc[ig['ctry'].isin(sample)]

import matplotlib as mpl
label_size = 8
mpl.rcParams['xtick.labelsize'] = label_size 
mpl.rcParams['ytick.labelsize'] = label_size 
mpl.rcParams['axes.labelsize'] = label_size 

 
fig, axes = plt.subplots(nrows=2, ncols=2, sharey = False)

sns.factorplot(data = ig, x="year", y = "Intermediate goods_tariff", hue="ctry",  ax=axes[0,0])
sns.factorplot(data = ig, x="year", y = "Capital goods_tariff", hue="ctry",  ax=axes[0,1])
sns.factorplot(data = ig, x="year", y = "Consumer goods_tariff", hue="ctry", ax=axes[1,0])
sns.factorplot(data = ig, x="year", y = "Raw materials_tariff", hue="ctry",  ax=axes[1,1])

plt.tight_layout()
```
